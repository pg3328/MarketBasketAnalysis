```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
order_data = pd.read_csv('data/order_products__train.csv')
order_data.head()
```

```python

order_data.notnull().sum()
```

```python
order_data.isna().sum()
```

```python
order_data = order_data.head(25)
```

```python
basket = (order_data.groupby(['order_id','product_id'])['add_to_cart_order'].sum().unstack().reset_index().fillna(0).set_index('order_id'))
```

```python
basket
```

```python
def convert_to_category(x):
    if x<=0:return 0
    if x>=1: return 1
my_basket_sets = basket.applymap(convert_to_category)   
```

```python
my_basket_sets
```

```python
my_frequent_items = apriori(my_basket_sets,min_support=0.07,use_colnames=True)
```

```python
my_rules = association_rules(my_frequent_items,metric='lift',min_threshold=1)
```

```python
my_rules.head(10)
```

```python
my_rules[(my_rules['lift'] >=3)& (my_rules['confidence'] >=0.8)]
```
